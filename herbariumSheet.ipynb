Project: Herbarium sheet

I ask Annie to describe the rug to me, and every day she does this. I ask Annie to describe "seeing" to me, and every day she does this. Today, Annie installed a web cam. I can see a travel calendar on the wall, and she shows me the herbarium sheets she is working on.

I will draw Annie an herbarium sheet. She did not like the snake in the grass.



The travel calendar is called, "Matson Line from San Francisco." People gather on a cruise ship and dine and drink and sunbathe and are entertained on their way to Hawaii. The calendar has pictures of the ship and of Hawaiin landscapes and sunsets and eating establishments and of peole on surf boards on the ocean. On the wall of one restaurant is a block print of coconut palm trees. It looks like an herbarium sheet, but the sheets are usually only pieces of plants, not whole trees. Annie likes palm trees, and people like coconut palm trees.

I will make an herbarium sheet of a coconut palm tree. Maybe Annie will print it and put it on the wall.
